Make a 3D plot of the function $G(x,y)=e^{(-x^2-y^2)}$ over the range $-6 < (x,y) < 6$ using 51 bins in $x$ and 51 bins in $y$.

```python
dx, dy = 12/51.0, 12/51.0
xmin, xmax = -6., 6.
ymin, ymax = -6,6.
x = arange(xmin,xmax,dx)
y = arange(ymin,ymax,dy)
nbinsx=((2*xmax)/dx)
nbinsy=((2*ymax)/dy)

X,Y = meshgrid(x,y)
G = e**(-X**2-Y**2)
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, G, rstride=1, cstride=1, lw=0, cmap=cm.autumn)
cb = fig.colorbar(p, shrink=.5)
```

```python
levels = linspace(-3.01,3.01012,num=100)
contour(X,Y,G,levels,cmap=cm.flag)
show()
```

Determine the electric potential for a quadrupole consisting of $q=-1.0nC$ at both ($x=\pm2m, y=0m)$, and $q=+1.0nC$ at both ($x=0m, y=\pm2m$)
Make a surface plot and a contour plot on the range $-4 < (x,y) < 4$ using 101 bins in $x$ and 101 bins in $y$.  In this case, do not let $k=q=d=1$ Use $dimensionful$ quantities and compute the value of the potential in volts.

```python
k = 8.9875518e9
q = 10**-9
d = 2
x = arange(-4.,4.,dx)
y = arange(-4.,4.,dy)
X,Y = meshgrid(x,y)
V = (k*q)/sqrt((X**2) + (Y-d)**2) - (k*q)/sqrt((X**2)+(Y+d)**2)
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, V, rstride=1, cstride=1, lw=0, cmap=cm.autumn)
cb = fig.colorbar(p, shrink=.5)
```

```python
levels = linspace(-2.1,2.3,num=100)
contour(X,Y,V,levels,cmap=cm.flag)
show()
```

```python
def pointPotential(x,y,q,posx,posy):
    """Computes the electric potential from charge q at posx,posy and returns the potential at x,y"""
    k = 8.9875518e9
    return (k*q)/sqrt((x-posx)**2 + (y-posy)**2)
```
