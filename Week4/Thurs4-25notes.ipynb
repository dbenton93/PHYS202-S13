```python
from Electrostatics import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
x,y = meshgrid(arange(-3.,3.,.037),arange(-3.,3.,.037))
Vdipole = dipolePotential(x,y,3e-9,.01)
Vpoint = pointPotential(x,y,-2e-9,0,-.005)
Vtotal = Vdipole + Vpoint


fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(x,y, Vtotal, rstride=1, cstride=1, lw=0, cmap=cm.PRGn)
cb = fig.colorbar(p, shrink=.5)
```
