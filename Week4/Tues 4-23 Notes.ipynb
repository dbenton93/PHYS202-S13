Electric Dipole Potential

'$V(x,y) = \cfrac{kq}{\sqrt{x^2 + (y-d)^2}} - \cfrac{kq}{\sqrt{x^2+(y + d)^2}}$

```python
k = 1
q=1
d=1
Vxy = (k*q)/(sqrt(x**2+(y-d)**2)) - (k*q)/(sqrt(x**2+(y+d)**2))
```

```python
print Vxy
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = fig.gca(projection ='3d')
p = ax.plot_surface(x, y, Vxy, rstride=1, cstride=1, linewidth=0, cmap=cm.Accent_r)
cb = fig.colorbar(p, shrink=0.5)
cset = ax.contour(x,y,Vxy, zdir='z', offset = -40., cmap=cm.rainbow)

```

```python
levels = array([-8,-4,-2,-1,-0.5,-0.25,0,.25,0.5,1,2,4,8])
fig = figure()
contour(x, y, Vxy, levels, cmap=cm.Accent_r)
show()
```
