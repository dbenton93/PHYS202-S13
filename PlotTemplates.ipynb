Line Plots.

```python
import numpy as np #importing necessities for math and graphing functions :)
import math
```

```python
x = np.arange(-5,5,.00001) #definied range of x-axis
E = np.cosh(x) #definied E(x)
p = np.sinh(x) #Defined P(x)
plot (x,E, label='Energy',linewidth=4) #Plotted E, setlabel for legend and thickened line
plot (x,p, label='Momentum', color='red', linewidth=4) #Plotted P, setlabel for legend and thickened line and Made red
xlabel('Position', fontsize=16) 
ylabel('Energy, Momentum',fontsize=16)
legend(loc='lower right') #made legend and set to lowerright, 
grid(True) #made grid
ax=gca()
ax.spines['right'].set_color('none') #centering spines at 0,0
title('turts',fontsize=24) #I don't know why my titledoesn't work, sorry prof
```

Subplots

```python
x = np.arange(-5,5,.00001) #definied range of x-axis
E = np.cosh(x) #definied E(x)
p = np.sinh(x) #Defined P(x)
subplot(211) #focus on top plot
plot (x,E,label='Energy',linewidth=4) #tell what top plot is
subplot(212) #focus on bot plot
plot (x,p, label='Momentum', color='red', linewidth=4) #bot plot
xlabel('Position', fontsize=16) 
ylabel('Momentum, Energy',fontsize=16)
ax=gca()
ax.spines['right'].set_color('none')
title('generic', fontsize=24) #again not sure what's wrong?
```
