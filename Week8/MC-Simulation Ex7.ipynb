Exercise 2. Exponential function - numerical
--------------------------------------------

```python
x = arange(0.,5.1,0.1)
lam = 1
f = lambda x: lam*exp(-lam*x)
Fx = cumsum(f(x)) / sum(f(x))
```

```python
N = 10000
z = numpy.random.random_sample(N)
xsamples = x[searchsorted(Fx,z)]
hist(xsamples, 50, normed=True, label="Sampled distribution")
plot(x,f(x),label='True distribution', lw=3, color='red')
legend(loc=0)
show()
```

```python
def distribute1D(x,prob,N):
    """Takes any numerical distribution prob, on the interval defined by the array x, and returns an array of N sampled values that are statistically the same as the input data"""
    Fx = cumsum(prob)/sum(prob)
    z = numpy.random.random_sample(N)
    xsamples = x[searchsorted(Fx,z)]
    return xsamples
```

```python
def distribute1D(x,prob,N):
    """Takes any numerical distribution prob, on the interval defined by the array x, and returns an array of N sampled values that are statistically the same as the input data"""
    Fx = cumsum(prob)/sum(prob)
    z = numpy.random.random_sample(N)
    Finv = x[searchsorted(Fx,z)]
    return Fx,Finv
```

```python
N=10000
x = arange(0.,1.01,.01)
f = lambda x: 60*x**3*(1-x)**2
Fx,Finv = distribute1D(x,f(x),N)
hist(Finv,50,normed=True,label="Sampled Distribution")
plot(x,f(x),'r-',linewidth=2.,label="True Distribution")
plot(x,Fx,'k-',linewidth=2.,label="CDF")
legend(loc=0)
show()
```
