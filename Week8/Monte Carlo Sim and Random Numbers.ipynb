SAMPLING FROM A CONTINUOUS DISTRIBUTION
---------------------------------------

```python
x = arange(0.,pi+.001,.01)
f = lambda x: sin(x)/2.
F = lambda x: (cos(0)-cos(x))/2.

plot(x,f(x),label="True Distribution",lw=3)
plot(x,F(x),label="Cumulative Function",lw=3)
xlim(0,3.5)
ylim(0,1.1)
legend(loc=0)
show()
```

```python
N = 10000
z = numpy.random.random_sample(N);
Finv = lambda z: arccos(cos(0)-2*z)
xsamp = Finv(z);
hist(xsamp,50,normed=True,label="Sampled Distribution")
plot(x,f(x),label='True Distribution',lw=3,color='red')
legend(loc=0)
show()
```

Exercise 1. Exponential function - analytic
-------------------------------------------

```python
lam = 1
x = arange(0,5.0001,.0001)
f = lambda x: lam*exp(-1*lam*x)
F = lambda x: 1-exp(-1*lam*x)
z = numpy.random.random_sample(N);
Finv = lambda z: -log(1-z)/lam
xsamp = Finv(z);
hist(xsamp,50,normed=True,label='Sampled Distribution')
plot (x,f(x), label='True Distribution', lw=3,color='red')
legend(loc=0)
show()
```
