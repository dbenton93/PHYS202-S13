Create a very simple plot

```python
import numpy as np
t = np.arange(0.0, 1.01, .01)
c = np.cos(4*np.pi*t)
s = np.sin(4*np.pi*t)
```

```python
fig = plot(t, c, color="red", linewidth=4, label="$cos(4{\pi}t)$", linestyle=":")
fig = plot(t, s, color="blue", linewidth=4, label="$sin(4{\pi}t)$", linestyle = "--")
xlabel('time (s)')
ylabel('Voltage (mV)')
title("dat plot", fontsize=72)
grid(True)
savefig("Matplotdatplot.pdf")
legend(loc="lower left")
```
