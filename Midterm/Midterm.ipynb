```python
def is_anagram(s1, s2):
    if len(s1) == len(s2):
        for i in s1:
            if s1[i] in s2 == True:
                return true
            else:
                return false
    else:
        return false
```

```python
s1 = 'cat'
s2 = 'act'
is_anagram(s1, s2)
```
