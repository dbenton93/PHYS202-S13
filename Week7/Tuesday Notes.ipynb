ODE's

(a) approximate solution

```python
g = 9.8
l = 0.5
th0 = 20.*pi/180
t=arange(0.,10.01,.01)
theta_app = lambda t: th0*cos(sqrt(g/l)*t)
plot(t,theta_app(t))
show()
```

```python
def deriv(theta_c,t):
    g = 9.8
    l=0.5
    theta = theta_c[0]
    omega = theta_c[1]
    dtheta = omega
    domega = -(g/l)*sin(theta)
    return array([dtheta,domega])
```

```python
from scipy.integrate import odeint
```

```python
t = arange(0.,10.01,0.01)
omega0 = 0.
theta_init = array([th0, omega0])

theta = odeint(deriv,theta_init,t)
```

```python
theta.shape
```

```python
plot(t,theta_app(t), 'b-', t,theta[:,0],'r--')
show()
```
